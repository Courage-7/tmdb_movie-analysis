### Research Question 1: How the runtime of movies has changed over the years? 

According to the Academy of Motion Picture Arts and Science, the American Film institute and British Film institute," a film/movie runs for more than 40mins",while Screen Actors Guild asserts that "a features running time is 60mins or longer".

So we will be filtering-out movies with a runtime greater than 40mins and then detailed comparisons of runtime of movies over the years.

```python

#which movies has runtime shorter than 40 minutes ?
df_movie_new[['original_title', 'runtime']]
print("List of movies with runtime shorter than 40mins")
df_movie_new[(df_movie_new['runtime'] <= 40) & (df_movie_new['runtime'] >=0)]
```

    This code returned a dataframe of 251 rows by 15 columns of movies with runtime greater than 40mins.

Plotting a histogram of movies runtime ( > 40) against the number of movies
creating and returning a function that ouputs a histogram of no. movies with the highest runtime after being called.

```python
def plotter():
    plt.hist(df_movie_new['runtime'], range = (40, 200))
    plt.title('Movies Runtime', fontsize = 23, color = 'green')
    plt.xlabel('Minutes', fontsize = 15, color = 'red')
    plt.ylabel('Number of movies', fontsize = 15, color = 'red')
    plt.show()
    return plotter
```

```python
plotter()
```

   From the histogram, the majority of movies released in our dataset had a runtime of 90 to 120mins.
    This happened as we called the potter() function created to create this visual.
